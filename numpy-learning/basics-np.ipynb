```python
arr5 = np.array([1,2,3])
arr6 = np.array([[1,2,3], [4,5,6]])
print(arr5.shape)
print(arr6.shape)
print(arr5.ndim, arr6.ndim)
```

Broadcasting is NumPy's way of doing arithmetic between arrays of different shapes — without manually resizing them.
Simple definition — NumPy automatically stretches the smaller array to match the shape of the larger one so the operation can work.
